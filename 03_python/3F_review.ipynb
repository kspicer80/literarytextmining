### 1. Which of the original seven novels mentions Harry the most (and least)? And which Hermione and Ron?

```python
# First let's define a list of all of the files

filenames = ['../workspace/classwork_425/potter_orig/Sorcerers Stone.txt',
             '../workspace/classwork_425/potter_orig/Chamber of Secrets.txt',
             '../workspace/classwork_425/potter_orig/Prisoner of Azkaban.txt',
             '../workspace/classwork_425/potter_orig/Goblet of Fire.txt',
             '../workspace/classwork_425/potter_orig/Order of the Phoenix.txt',
             '../workspace/classwork_425/potter_orig/Half-Blood Prince.txt',
             '../workspace/classwork_425/potter_orig/Deathly Hallows.txt',
            ]
```

```python
# Let's define a list of the three characters
trio = ['Harry', 'Hermione', 'Ron']
```

```python
# Then let's loop over the files
for filename in filenames:
    
    # Let's open the file
    with open(filename) as file:
        text = file.read()
    
    # Let's tokenize the file
    import nltk
    tokens = nltk.word_tokenize(text)
    
    ### Let's make a dictionary of counts
    # Make an empty dictionary
    counts = {}
    
    # For each token in the list of tokens...
    for token in tokens:
        # (optional) don't bother counting words that are not names of characters in the trio
        if token not in trio:      
            continue   
            
        # if the token is in the dictionary
        if token not in counts:
            # set it to 1
            counts[token]=1
        # otherwise
        else:
            # add 1 to it
            counts[token]+=1
    ###
    
    ###
    # Now let's print how often Harry, Hermione, and Ron appear
    # Print the filename (so we know where we are in the loop)
    print('##',filename)
    
    # For each character in the trio...
    for character in trio:
        # the count of that character is stored in the dictinonary 'counts'
        count = counts.get(character,0)    # get the count; but if it's not there, return 0
        
        # make a normalized count
        freq = count / len(tokens)
        perc = round(freq*100,2)
        
        # print
        print(character, count, perc)
    
    # print an empty line
    print()
    ###
```

```python
# Let's loop over the files

for filename in filenames:
    # Let's open the file
    with open(filename) as file:
        text = file.read()
    
    # Let's tokenize the file
    import nltk
    tokens = nltk.word_tokenize(text)
    
    # Let's put it into a dictionary
    from collections import Counter
    counts = Counter(tokens)
    
    # Count the number of women name mentions
    count_women = 0
    for woman in women:
        count_women += counts.get(woman,0)
    
    # Count the number of men name mentions
    count_men = 0
    for man in men:
        count_men += counts.get(man,0)
    
    # Let's also make a normalized version
    num_tokens = len(tokens)
    perc_women = round(count_women/num_tokens*100,2)
    perc_men = round(count_men/num_tokens*100,2)
    
    # Now let's print how often Harry, Hermione, and Ron appear
    print('##',filename)
    print('Women', count_women, perc_women)
    print('Men', count_men, perc_men)
    print('Men/Women', perc_men / perc_women)
    
    # print an empty line
    print()
```
