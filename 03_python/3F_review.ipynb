```python
# # First let's define a list of all of the files

filenames = ['../corpora/harry_potter/texts/Sorcerers Stone.txt',
             'classwork_425/potter_orig/Chamber of Secrets.txt',
             'classwork_425/potter_orig/Prisoner of Azkaban.txt',
             'classwork_425/potter_orig/Goblet of Fire.txt',
             'classwork_425/potter_orig/Order of the Phoenix.txt',
             'classwork_425/potter_orig/Half-Blood Prince.txt',
             'classwork_425/potter_orig/Deathly Hallows.txt',
            ]
```

```python
# Let's define a list of the three characters
trio = ['Harry', 'Hermione', 'Ron']
```

```python
def count_tokens(tokens):
    # from tokens to counts
    counts = {}
    
    # for each token in the list of tokens...
    for token in tokens:
        # if the token is not in the dictionary of counts...
        if token not in counts:
            counts[token]=1
        # otherwise...
        else:
            counts[token]+=1
            
    return counts
```

```python
def count_tokens2(tokens):
    from collections import Counter
    counts = Counter(tokens)
    return counts
```

```python
# For each file in the list of filenames
for filename in filenames:
    print(filename)
    
    # open the file
    with open(filename) as file:
        text_str = file.read()
        
    # tokenize
    import nltk
    tokens = nltk.word_tokenize(text_str.lower())
    
    # count
    counts = count_tokens(tokens)
            
    # answer the Q
    for char in trio:
        count = counts.get(char.lower(),0)
        tf = round(count / len(tokens),4)
        print(char,count,tf)
    
    print()
```

```python
women="""Hermione
Ginny
Luna
Nymphadora
Lily
Bellatrix
Narcissa
Petunia
Fleur
Cho
Helga
Rowena
Nagini
Lavender
Pansy
Parvati
Padme
Molly
Penelope
Dolores
Minerva
Madame
Sybill
Rita""".split('\n')
```

```python
men="""Ron
Severus
Draco
Albus
Neville
Kingsley
James
Sirius
Remus
Cedric
Gellert
Tom
Peter
Viktor
Dudley
Blaise
Godric
Salazar
Vernon
Hedwig
Newt
Lee
Fred
George
Bill
Charlie
Percy
Arthur
Oliver
Cornelius
Mad-Eye
Cornelius
GIlderoy
Barty
Lucius
Hagrid""".split('\n')
```

```python
print(len(women), len(men))
```

```python
def tokenize(string):
    import nltk
    
    # make lowercase
    string = string.lower()
    
    # tokenize
    tokens = nltk.word_tokenize(string)
    
    return tokens
```

```python
# for every file...
for filename in filenames:
    
    # open
    with open(filename) as file:
        text_str = file.read()
        
    # tokenize
    tokens = tokenize(text_str)
    
    # count
    counts = count_tokens2(tokens)
    
    # defaults
    count_forallmen = 0
    count_forallwomen = 0
    
    # loop through the women
    
    # for each individual woman, in the list of women...
    for woman in women:
        # get the count for that particular woman
        count_forthiswoman = counts.get(woman.lower(),0)
        
        # add that count to the sum count for all women
        count_forallwomen += count_forthiswoman
        
    # for each individual man, in the list of men...
    for man in men:
        # get that person's count
        count_thisman = counts.get(man.lower(),0)
        # add that count to the total count for men
        count_forallmen += count_thisman
        
    # get the ratio between the count for all men and the count for all women
    mf_ratio = count_forallmen / count_forallwomen
        
    
    print(filename,count_forallwomen,count_forallmen, mf_ratio)
        
```
